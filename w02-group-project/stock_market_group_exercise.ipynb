```python
# let's start by looking at the data in its rawness

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(9, 5))
image_1 = plt.imread('S&P500_tableau.png')
plt.imshow(image_1)
plt.show()

plt.figure(figsize=(9, 5))
image_2 = plt.imread('Eurofirst_tableau.png')
plt.imshow(image_2)
plt.show()

plt.figure(figsize=(5, 10))
image_3 = plt.imread('Bovespa_tableau.png')
plt.imshow(image_3)
plt.show()

plt.figure(figsize=(9, 6))
image_4 = plt.imread('HangSeng_tableau.png')
plt.imshow(image_4)
plt.show()
```
