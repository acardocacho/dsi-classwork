## Merge all the dataframes into giant one, indexed by date

```python
dfs = [sp_df, euronext_df, brazil_df, hongkong_df]
```

```python
for df in dfs:
    print(df.columns)

for df in dfs:
    df["Date"] = pd.to_datetime(df["Date"])
    df.index = df["Date"]
    del df["Date"]
    print(df.dtypes)

for df in dfs:
    print(df.columns)
```

### Rename the columns to reflect the dataset they came from

```python
# dfs = [sp_df, euronext_df, brazil_df, hongkong_df]
dfs_names = ["SP", "EU", "BR", "HK"]
dfs_columns = ["Open", "High", "Low", "Close", "Volume", "Adj Close"]

col_names = []
for x in dfs_names:
    for y in dfs_columns:
        col_names.append(x+"-"+y)
col_names
```

```python
all_df = pd.concat(dfs, axis=1)
all_df.head()
```

```python
all_df.columns = col_names
all_df.head()
```

```python
all_df.describe()
```

```python
all_df.info()
#all_df.rename(columns =)
```

```python
all_df.to_csv("all.csv")
```

## Make one giant plot

```python
all_df[["SP-Close", "EU-Close", "BR-Close", "HK-Close"]].resample("D").mean().plot(figsize=(12, 6), logy=True)
#plt.yscale("log")
#plt.show()
```

## And now some pivot tables

```python
day_pivot = all_df.pivot_table(index=all_df.index.day, 
                               values=['SP-Open', 'SP-High', 'SP-Low', 'SP-Close', 'SP-Volume'], 
                               aggfunc=np.mean)
day_pivot
```

```python
day_pivot["SP-Volume"].plot()
```

```python
month_pivot = all_df.pivot_table(index=all_df.index.month, 
                               values=['SP-Open', 'SP-High', 'SP-Low', 'SP-Close', 'SP-Volume'], 
                               aggfunc=np.mean)
month_pivot
```

```python
month_pivot["SP-Volume"].plot()
```

```python
weekday_pivot = all_df.pivot_table(index=all_df.index.weekday, 
                               values=['SP-Open', 'SP-High', 'SP-Low', 'SP-Close', 'SP-Volume'], 
                               aggfunc=np.mean)
weekday_pivot
```

```python
weekday_pivot["SP-Volume"].plot(kind="bar")
pylab.xticks([0,1,2,3,4], ["Mon", "Tue", "Wed", "Thu", "Fri"])
pylab.show()
```
